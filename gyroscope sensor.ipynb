```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
acc = pd.read_csv('Gyroscope.csv')
```

```python
acc.head()
```

```python
acc = acc.rename(columns={'wx (rad/s)' : 'wx','wy (rad/s)' : 'wy'
                    ,'wz (rad/s)' : 'wz'})

acc.head()
```

```python
#x axis

plt.title('gyroscope analysis of x y z - direction' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['wx'], linewidth=1, label='X axis velocity')
plt.plot(acc['time'], acc['wy'], linewidth=1, label='Y axis velocity')
plt.plot(acc['time'], acc['wz'], linewidth=1, label='Z axiis velocity')

plt.xlabel('time (s) --->')
plt.ylabel('angular velocity (rad/s)  -->')

plt.legend()
plt.show()
```
