```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
acc = pd.read_csv('magnetometer.csv')
```

```python
acc.head()
```

```python
# acc = acc.rename(columns={'wx (rad/s)' : 'wx','wy (rad/s)' : 'wy' ,'wz (rad/s)' : 'wz'})

# acc.head()
```

```python
plt.title('magnetic field analysis of x y and z direction' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['Bx'], linewidth=1, label='X-axis magnetic field')
plt.plot(acc['time'], acc['By'], linewidth=1, label='Y-axis magnetic field')
plt.plot(acc['time'], acc['Bz'], linewidth=1, label='Z-axis magnetic field')

plt.xlabel('time (s) --->')
plt.ylabel('magnetic field (uT)  -->')

plt.legend()
plt.show()
```

```python
plt.title('magnetometer analysis of absolute magenetic field' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['BT'], linewidth=1, label='magnetic field')

plt.xlabel('time (s) --->')
plt.ylabel('absolute magnetic feild (uT)  -->')

plt.legend()
plt.show()
```
