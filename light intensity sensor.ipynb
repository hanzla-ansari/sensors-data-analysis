```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python

acc = pd.read_csv('light.csv')

plt.title('light sensor analysis with time' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['I'], linewidth=1, label='X')

plt.xlabel('time (s) --->')
plt.ylabel('Illuminance (lx)  -->')

plt.legend()
plt.show()
```

```python
acc['time'] = acc['time']*1.34
acc = acc.rename(columns={'time' : 'distance'})
# x- axis
plt.title('light sensor analysis with time' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['distance'], acc['I'], linewidth=1, label='X')

plt.xlabel('disatance (m) --->')
plt.ylabel('Illuminance (lx)  -->')

plt.legend()
plt.show()
```

```python
# acc = acc.rename(columns={'time' : 'distance'})

gps = pd.read_csv('gps.csv')
gps = gps.rename(columns={'Latitude ()' : 'LATITUDE', 'Longitude ()':'LONGITUDE'})

gps = gps.loc[1:, ['LATITUDE','LONGITUDE']]
# # x- axis
plt.figure(figsize=(8,4))
plt.title('graph' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(gps['LONGITUDE'],gps['LATITUDE'],  linewidth=5, label='travelled path')

plt.xlabel('LONGITUDE --->')
plt.ylabel('LATITUDE  -->')

plt.legend()
plt.show()
```
