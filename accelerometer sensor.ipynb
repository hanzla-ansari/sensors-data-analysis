```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
acc = pd.read_csv('accelerometer.csv')
```

```python
acc = acc.rename(columns={'ax (m/s^2)' : 'ax','ay (m/s^2)' : 'ay'
                    ,'az (m/s^2)' : 'az', 'aT (m/s^2)' : 'aT'})

acc.head()
```

```python
#x axis

plt.title('Accerometer analysis of x-direction' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['ax'], linewidth=1, label='X')

plt.xlabel('time (s) --->')
plt.ylabel('acceleration (m/s^2)  -->')

plt.legend()
plt.show()
```

```python
#y-axis

plt.title('Accerometer analysis of y-direction' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['ay'], linewidth=1, label='Y', color='r')

plt.xlabel('time (s) --->')
plt.ylabel('acceleration (m/s^2)  -->')

plt.legend()
plt.show()
```

```python
#z-axis

plt.title('Accelerometer analysis of z-direction' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['az'], linewidth=1, label='Normal activity', color='y')

plt.xlabel('time (s) --->')
plt.ylabel('acceleration (m/s^2)  -->')

plt.legend()
plt.show()
```

```python
#all in one 

plt.title('Accelerometer analysis of xyz-direction' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})


plt.plot(acc['time'], acc['az'], linewidth=1, label='Z', color='g')
plt.plot(acc['time'], acc['ay'], linewidth=1, label='Y', color='r')
plt.plot(acc['time'], acc['ax'], linewidth=1, label='X')

plt.xlabel('time (s) --->')
plt.ylabel('acceleration (m/s^2)  -->')


plt.legend()
plt.show()
```

```python
#all in one 

plt.title('Accelerometer analysis of absolute acceleration' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(acc['time'], acc['aT'], linewidth=1, label='abs', color='#ababab')

plt.xlabel('time (s) --->')
plt.ylabel('acceleration (m/s^2)  -->')


plt.legend()
plt.show()
```

```python
jumps = acc.loc[acc['aT']>4]
rest =  acc.loc[acc['aT']<=4]
```

```python
#all in one 

plt.title('Accerometer analysis of jumps activity' , fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.plot(rest['time'], rest['aT'], linewidth=1, label='walking/standing/sitting', color='#ababab')
plt.plot(jumps['time'], jumps['aT'], 
         linewidth=1, label='Jumping/running upstrairs', color='g', linestyle='-.')

plt.xlabel('time (s) --->')
plt.ylabel('acceleration (m/s^2)  -->')


plt.legend()
plt.savefig('jumps activity')
plt.show()

```
